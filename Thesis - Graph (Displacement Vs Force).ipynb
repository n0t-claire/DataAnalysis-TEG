```python
FORCE_COL = 'Force_N'
DISPLACEMENT_COL = 'Displacement_mm'
FILE_PATH = 'Combined mUTM Data.xlsx'

Sheet_const = {
    'TE Fins':{'A0': 40500, 'L0': 150},
    'Polyimide':{'A0': 36, 'L0': 12},
    '12.5% PVA':{'A0': 15.6913, 'L0':14.01},
    '10.5% PVA': {'A0': 15.6913, 'L0':14.01}
}

try:
    all_sheets = pd.read_excel(FILE_PATH, sheet_name=None)

except FileNotFoundError:
    print(f"Error: File not found at '{FILE_PATH}'. Exiting.")
    exit()

all_data_combined = pd.DataFrame()

for sheet_name, df in all_sheets.items():
    if sheet_name not in Sheet_const:
        print(f"WARNING: Skipping sheet '{sheet_name}'. Constants missing.")
        continue
    
    # --- Data Cleaning and Validation (CRITICAL STEP) ---
    df = df.copy()
    
    # Force columns to numeric, converting errors to NaN
    df[FORCE_COL] = pd.to_numeric(df[FORCE_COL], errors='coerce')
    df[DISPLACEMENT_COL] = pd.to_numeric(df[DISPLACEMENT_COL], errors='coerce')
    
    # Drop rows where Force or Displacement are not valid numbers
    df.dropna(subset=[FORCE_COL, DISPLACEMENT_COL], inplace=True)
    
    # Skip sheet if there are not enough rows left
    if df.empty or len(df) < 5:
        print(f"WARNING: Sheet '{sheet_name}' has too few valid data points. Skipping.")
        continue
    
    # --- Calculation ---
    const = Sheet_const[sheet_name]
    df['Test_Source'] = sheet_name
    df['Stress_MPa'] = df[FORCE_COL] / const['A0']
    df['Strain'] = df[DISPLACEMENT_COL] / const['L0']
    
    all_data_combined = pd.concat([all_data_combined, df], ignore_index=True)
# --- DEBUG: Print Max Stress and Max Force Values ---

if all_data_combined.empty:
    print("Error: The combined DataFrame is empty.")
else:
    print("\n--- DEBUG: Maximum Calculated Stress (MPa) ---")
    max_stress = all_data_combined.groupby('Test_Source')['Stress_MPa'].max()
    print(max_stress)
    print("-" * 40)

    print("\n--- DEBUG: Maximum Raw Force (N) ---")
    max_force = all_data_combined.groupby('Test_Source')[FORCE_COL].max()
    print(max_force)
    print("-" * 40)
    
# --- Plotting Section ---
if all_data_combined.empty:
    print("Error: The combined DataFrame is empty even after cleaning. Cannot plot.")
else:
    # 5a. Violin Plot (Check the distribution)
    plt.figure(figsize=(10,6)) 
    sns.violinplot(
        x="Test_Source", 
        y="Stress_MPa", 
        data=all_data_combined, 
        palette="viridis", 
        inner="quartile", # Using 'quartile' as requested
        cut=0
    )
plt.title('Distribution of Stress Across TE Components (Cleaned Data)')
plt.xlabel('Test Groups')
plt.ylabel('Stress (MPa)')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
